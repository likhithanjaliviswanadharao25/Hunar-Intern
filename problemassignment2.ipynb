```python
import pandas as pd
import matplotlib.pyplot as plt

# === Historical Height Dataset of UK Males ===

# Prepare the dataset: years and corresponding average male heights in cm
years = [1810, 1820, 1830, 1840, 1850, 1860, 1870, 1880, 1900, 1910,
         1920, 1930, 1940, 1950, 1960, 1970, 1980]
heights_cm = [169.7, 169.1, 166.7, 166.5, 165.6, 166.6, 167.2, 168,
              169.4, 170.9, 171, 173.9, 174.9, 176, 176.9, 177.1, 176.8]

# Create a DataFrame from the lists
height_data = pd.DataFrame({
    "Year": years,
    "Height_cm": heights_cm
})

# Compute average height over the years
average_height = height_data["Height_cm"].mean()
print(f"Average male height (1810â€“1980): {average_height:.2f} cm")

# === Plotting the trend of height over the years ===

plt.figure(figsize=(10, 6))  # Set figure size
plt.plot(height_data["Year"], height_data["Height_cm"],
         linestyle='-', marker='o', linewidth=2)

plt.title("Trend of Male Height in the UK (1810â€“1980)")
plt.xlabel("Year")
plt.ylabel("Height in Centimeters")
plt.grid(True)  # Show grid lines
plt.tight_layout()
plt.show()
```
